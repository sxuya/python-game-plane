## 建立飞机

```python
import random
import pygame

background = 'cell.jpg'
plane = 'feiji.png'
enemy = 'feiji2.png'

ex1 = random.randrange(20, 200)
ex2 = random.randrange(200, 400)
ex3 = random.randrange(400, 600)

ey1 = random.randrange(10, 50)
ey2 = random.randrange(10, 50)
ey3 = random.randrange(10, 50)


pygame.init()
pygame.display.set_caption('Python App')
screen = pygame.display.set_mode([640, 320])

bg = pygame.image.load(background).convert()
pln = pygame.image.load(plane).convert_alpha()
pln = pygame.transform.scale(pln, (80, 80))
pln = pygame.transform.rotate(pln, 90)

enm = pygame.image.load(enemy).convert_alpha()
enm = pygame.transform.scale(enm, (80, 80))
enm = pygame.transform.rotate(enm, -90)

y_move = 0
x_color = 0
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
            
    screen.blit(bg, (0, y_move))
    y_move -= 0
    
#     screen.blit(pln, (40, 240))
    screen.blit(enm, (ex1, ey1))
    screen.blit(enm, (ex2, ey2))
    screen.blit(enm, (ex3, ey3))
    ey1 += random.randrange(1, 8)/100
    ey2 += random.randrange(1, 8)/100
    ey3 += random.randrange(1, 8)/100
    
    
    pygame.display.update()
    
    x_color += 0.5
    screen.fill([(x_color+150)%255, (x_color+100)%250, x_color%250])
```

## 建立墙体

```python
import random
import pygame

w1 = 'wall1.png'
w2 = 'wall2.png'
player = 'tree.png'

pygame.init()
pygame.display.set_caption('Python App')
screen = pygame.display.set_mode([640, 320])

w_1 = pygame.image.load(w1).convert_alpha()
w_1 = pygame.transform.scale(w_1, (80, 80))
# w_2 = pygame.image.load(w2).convert_alpha()
# w_2 = pygame.transform.scale(w_2, (80, 80))
plr = pygame.image.load(player).convert_alpha()
plr = pygame.transform.scale(plr, (160, 160))

total = 16
walls = []
while total > 0:
    walls.append(w_1)
#     if total % 2 == 0:
#         walls.append(w_1)
#     else:
#         walls.append(w_2)
    total -= 1
    

def blit_squences(data, x, y):
    if isinstance(data, list):
        for d in data:
            screen.blit(d, (x, y))
    else:
        screen.blit(data, (x, y))
        

y_move = 0
x_color = 0

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
            
    init_x = 600
    init_y = 40
    step_x = -50
    step_y = 9
            
    for w in walls:
        screen.blit(w, (init_x, init_y))
        init_x += step_x
        init_y += step_y
        
    screen.blit(plr, (62, 50))
    pygame.display.update()
    
#     x_color += 0.5
#     screen.fill([(x_color+150)%255, (x_color+100)%250, x_color%250])
```

```python
import random
import pygame

pygame.init()
pygame.display.set_caption('Python App')
screen = pygame.display.set_mode([640, 320])

pln = pygame.image.load('feiji.png').convert_alpha()
pln = pygame.transform.scale(pln, (80, 80))
pln = pygame.transform.rotate(pln, 90)  # 逆时针度数

pln2 = pygame.image.load('feiji.png').convert_alpha()
pln2 = pygame.transform.scale(pln2, (80, 80))
# pln = pygame.transform.rotate(pln, 60)

pln3 = pygame.image.load('feiji.png').convert_alpha()
pln3 = pygame.transform.scale(pln3, (80, 80))
pln3 = pygame.transform.rotozoom(pln3, 60, 2)

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        
    screen.blit(pln, (62, 50))
    screen.blit(pln2, (250, 50))
    screen.blit(pln3, (350, 100))
    pygame.display.update()
```

### 剪切函数 chop

```python
import random
import pygame

pygame.init()
pygame.display.set_caption('Python App')
screen = pygame.display.set_mode([640, 320])

pln = pygame.image.load('feiji.png').convert_alpha()
pln = pygame.transform.scale(pln, (80, 80))

pln2 = pygame.image.load('feiji.png').convert_alpha()
pln2 = pygame.transform.scale(pln2, (80, 80))
pln2 = pygame.transform.chop(pln2, [0, 0, 20, 40])  # [x, y, w, h]，x+w，y+h，为最终点，右下角矩形内容显示

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        
    screen.blit(pln, (0, 0))
    screen.blit(pln2, (200, 100))
    
    pygame.display.update()
```

## Sprit class

```python
import pygame

pygame.init()
pygame.display.set_caption('Python App')
screen = pygame.display.set_mode([1280, 640])

class explode(pygame.sprite.Sprite):
    def __init__(self, target, frame, single_w, single_h, pos=(0, 0)):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load(target).convert_alpha()
        self.main_image = self.image
        self.frame = frame
        self.rect = self.image.get_rect()
        self.count = 0
        self.single_w, self.single_h = single_w, single_h
        self.rect.topleft = pos
        
    def update(self):
        if self.count < self.frame-1:
            self.count += 1
        else:
            self.count = 0
            
        self.image = self.main_image.subsurface([self.count * self.single_w, 0, self.single_w, self.single_h])
        
exp = explode('explode.png', 3, 276, 276, (100, 100))
# explode.png.size = 830 * 278
group = pygame.sprite.Group()
group.add(exp)

fps = 6
fpsClock = pygame.time.Clock()

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        
    group.update()
    group.draw(screen)
    
    pygame.display.update()
    fpsClock.tick(fps)
    screen.fill([100, 100, 100])
```

## 碰撞检测

### 碰撞函数

```python
import pygame

background = 'cell.jpg'
plane = 'feiji.png'
enemy = 'wall2.png'
tree = 'tree.png'

pygame.init()
pygame.display.set_caption('Python App')
screen = pygame.display.set_mode([320, 640])

bg = pygame.image.load(background).convert()

pln = pygame.image.load(plane).convert_alpha()
pln = pygame.transform.scale(pln, (80, 80))
pln = pygame.transform.rotate(pln, 90)

enm = pygame.image.load(enemy).convert_alpha()
enm = pygame.transform.scale(enm, (60, 60))

tre = pygame.image.load(tree).convert_alpha()
tre = pygame.transform.scale(tre, (50, 50))

def collide(a, axy, b, bxy):
    a_x1, a_x2 = axy[0], axy[0]+a.get_width()
    a_y1, a_y2 = axy[1], axy[1]+a.get_height()
    b_x1, b_x2 = bxy[0], bxy[0]+b.get_width()
    b_y1, b_y2 = bxy[1], bxy[1]+b.get_height()
    ax, ay = range(a_x1, a_x2), range(a_y1, a_y2)
    bx, by = range(b_x1, b_x2), range(b_y1, b_y2)
    
    ct = 0
    for a in ax:
        if a in bx:
            ct = 1
            break
    for a in ay:
        if a in by:
            if ct == 1:
                return True
    return False
    
y1, y2 = 1, 1

fps = 18
fpsClock = pygame.time.Clock()

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
            
    screen.blit(bg, (0, 0))
    screen.blit(pln, (50, 250 + y1))
    screen.blit(enm, (50, 50 + y2))
    
#     print(collide(pln, (50, 450 + y1), enm, (50, 50 + y2)))
    if collide(pln, (50, 250 + y1), enm, (50, 50 + y2)):  # 不能使用 while，不然 loop 出不来，卡死在这里。
        screen.blit(tre, (65, 250 + y1))
        
    
    
    y1 -= 1
    y2 += 1

    pygame.display.update()
    fpsClock.tick(fps)
```

## 添加 UI

```python
import pygame

background = 'cell.jpg'
plane = 'feiji.png'
enemy = 'wall2.png'
tree = 'tree.png'

pygame.init()
pygame.display.set_caption('Python App')
screen = pygame.display.set_mode([640, 640])

bg = pygame.image.load(background).convert()

pln = pygame.image.load(plane).convert_alpha()
pln = pygame.transform.scale(pln, (80, 80))
pln = pygame.transform.rotate(pln, 90)

enm = pygame.image.load(enemy).convert_alpha()
enm = pygame.transform.scale(enm, (60, 60))

tre = pygame.image.load(tree).convert_alpha()
tre = pygame.transform.scale(tre, (50, 50))

the_rect = pygame.Rect(5, 10, 250, 80)  # 前面两个参数有什么用呢？好像是根据 标记1 位置进行重新确定位置，这里无所谓的样子。
block_surface = screen.subsurface(the_rect).copy()  ## screen 建立自图层，大小是上面的 rect；copy 拷贝，避免图层锁定情况。

def text_out(text):
    fnt = pygame.font.Font('freesansbold.ttf', 15)
#     fnt = pygame.font.SysFont('微软雅黑', 15)  # 失败，不知道 encoding 要怎么处理
    tsurf = fnt.render(text, True, (255, 255, 255))  # 返回的是一个文字图层 （内容， 抗锯齿， RGB）
    trect = tsurf.get_rect()
    trect.center = ((block_surface.get_width()/4),(block_surface.get_height()/2))
    block_surface.blit(tsurf, trect)

fps = 18
fpsClock = pygame.time.Clock()

y1, y2 = 1, 1

while True:   
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
#             sys.exit()
        
    screen.blit(bg, (0, 0))
    screen.blit(pln, (50, 250 + y1))
    screen.blit(enm, (50, 50 + y2))
    
    txt = "Pause"
#     txt = u"暂停"
    x, y = pygame.mouse.get_pos()
    if pygame.mouse.get_pressed()[0]:
        if x >= 200 and x <= 350 and y >= 100 and y <= 240:
#             txt = u"点击"
            txt = "Clicked"
    screen.blit(block_surface, (200, 100))  # 标记1 位置
    block_surface.fill([0, 25, 0])
    text_out(txt)
    
    y1 -= 1
    y2 += 1

    pygame.display.update()
    fpsClock.tick(fps)
```
